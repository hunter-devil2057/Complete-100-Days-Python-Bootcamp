Write a Python Program to Generate Number from 100 to 1 only odd numbers. 

```python
i=100       #need to start from 100 so initializing as 100
# Checking the value of i as we have to run until it's 1
while i>=1:
    if i%2 !=0: #checking the number, whether it's odd
        print(i)
    i-=1
print("Second Loop is Started from 1 to 100 only even number")
n=1
while n<=100:       #continuing the loop
    if n%2==0:
        print(n)
    n+=1
```
